```python
missing_na = pd.isnull(market).sum()
print(missing_na)
```

```python
#cleaning data item_weight
market['Item_Weight'] = market.groupby('Item_Type')['Item_Weight'].transform(
    lambda x: x.fillna(x.median())
)

#cleaning data outlet size
market['Outlet_Size'] = market.groupby('Outlet_Type')['Outlet_Size'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Medium")
)
```

```python
missing_na = pd.isnull(market).sum()
print(missing_na)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pilih hanya kolom numerik
numeric_cols = market.select_dtypes(include='number').columns

# Atur ukuran grid plot (misalnya 2 kolom)
n_cols = 2
n_rows = (len(numeric_cols) + 1) // n_cols

plt.figure(figsize=(12, 5 * n_rows))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot(x=market[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

```python
Q1 = market['Item_Outlet_Sales'].quantile(0.25)
Q3 = market['Item_Outlet_Sales'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Ambil baris yang merupakan outlier
outlier_outlet_sales = market[(market['Item_Outlet_Sales'] < lower_bound) | (market['Item_Outlet_Sales'] > upper_bound)]

print(outlier_outlet_sales)
print("Jumlah outlier:", len(outlier_outlet_sales))
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Langkah 1: Ganti nilai 0 dengan median per item (biar logis secara konteks)
market['Item_Visibility'] = market.groupby('Item_Type')['Item_Visibility'].transform(
    lambda x: x.replace(0, x.median())
)

# Langkah 2: Transformasi cube root (akar pangkat 3)
market['visibility_cbrt'] = np.cbrt(market['Item_Visibility'])

# Langkah 3a: Boxplot hasil transformasi
plt.figure(figsize=(8, 5))
sns.boxplot(x=market['visibility_cbrt'], color='lightgreen')
plt.title("Boxplot - Item Visibility (Cube Root Transformed)")
plt.xlabel("Cube Root of Item Visibility")
plt.show()

# Langkah 3b: Histogram hasil transformasi
plt.figure(figsize=(8, 5))
sns.histplot(market['visibility_cbrt'], kde=True, bins=50, color='green')
plt.title("Distribusi Item Visibility (Cube Root Transformed)")
plt.xlabel("Cube Root of Item Visibility")
plt.ylabel("Frekuensi")
plt.show()
```

```python
Q1 = market['visibility_cbrt'].quantile(0.25)
Q3 = market['visibility_cbrt'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Ambil baris yang merupakan outlier
outlier_visibility_cbrt = market[(market['visibility_cbrt'] < lower_bound) | (market['visibility_cbrt'] > upper_bound)]

print(outlier_visibility_cbrt)
print("Jumlah outlier:", len(outlier_visibility_cbrt))
```
